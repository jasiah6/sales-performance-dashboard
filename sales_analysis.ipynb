**Convert Data Types**

*Order Date and Ship Date are in object format. These should be converted to datetime format.*


```python
df['Order Date'] = pd.to_datetime(df['Order Date'], format="%d/%m/%Y")
df['Ship Date'] = pd.to_datetime(df['Ship Date'], format="%d/%m/%Y")
```

*Feature Engineering*

```python
df.info()
```

```python
df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month
df['Day'] = df['Order Date'].dt.day
```

```python
df['Delivery Time'] = (df['Ship Date'] - df['Order Date']).dt.days
```

```python
df[['Order Date', 'Ship Date']].head()
```

*Exploratory Data Analysis (EDA)*

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Sales by Region
sales_by_region = df.groupby('Region')['Sales'].sum()
sales_by_region.plot(kind='bar', figsize=(6, 3), title='Total Sales by Region')
plt.xlabel('Region')
plt.ylabel('Sales')
plt.show()
```

```python
#Sales by Category
sales_by_category = df.groupby('Category')['Sales'].sum()
sales_by_category.plot(kind='pie', figsize=(5, 5), title='Sales Distribution by Category', autopct='%1.1f%%')
plt.ylabel('')
plt.show()
```

```python
#Sales Trends Over Time (monthly or yearly)
sales_by_month = df.groupby(['Year', 'Month'])['Sales'].sum()
sales_by_month.plot(kind='line', figsize=(10, 4), title='Monthly Sales Trends')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()
```

```python
#Top Products by Sales
top_products = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)
top_products.plot(kind='bar', figsize=(12, 3), title='Top 10 Products by Sales')
plt.xlabel('Product')
plt.ylabel('Sales')
plt.show()
```
